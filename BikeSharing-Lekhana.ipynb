<h3>Step 3: Data Preparation</h3>

<h4> Converting Categorical Fields to Dummy Variables
Categorical variables like season and weathersit are numeric but represent categories, so we convert them to dummy variables to avoid misleading ordinal relationships.</h4?


```python
print(cycling['season'].unique())

# Define the mapping dictionary
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}

# Apply the mapping to the 'season' column using map
cycling['season'] = cycling['season'].map(season_map)

# Check the updated DataFrame
cycling.head()

```

```python
print(cycling['weathersit'].unique())
# Define the mapping dictionary
weather_map = {1: 'Clear', 2: 'Cloudy', 3: 'Light Snow'}

# Apply the mapping using replace
cycling['weathersit'] = cycling['weathersit'].map(weather_map)

# Check the result
cycling.head()
```

```python
# Check bike rentals by year
sns.boxplot(x='yr', y='cnt', data=cycling)
plt.title('Bike Rentals by Year (2018 vs 2019)')
plt.xlabel('Year (0 = 2018, 1 = 2019)')
plt.ylabel('Total Rentals (cnt)')
plt.show()
```

```python
# Drop irrelevant columns
cycling = cycling.drop(columns=['instant', 'dteday'])
cycling.head()
```

```python
# Create a 'weekend' column
cycling['weekend'] = cycling['weekday'].apply(lambda x: 1 if x in [0, 6] else 0)

# Check average rentals by weekend vs. weekday
sns.boxplot(x='weekend', y='cnt', data=cycling)
plt.title('Bike Rentals: Weekend vs Weekday')
plt.xlabel('Weekend (1 = Weekend, 0 = Weekday)')
plt.ylabel('Total Rentals (cnt)')
plt.show()
```

```python
#Monthly Trends in Bike Usage
monthly_avg = cycling.groupby('mnth')['cnt'].mean()

plt.plot(monthly_avg.index, monthly_avg.values, marker='o')
plt.title('Average Bike Rentals by Month')
plt.xlabel('Month')
plt.ylabel('Average Rentals')
plt.xticks(range(1, 13))
plt.grid()
plt.show()
```

```python
#Peak months: Assign 1 for peak months (e.g., May, June) and 0 otherwise.

cycling['peak_month'] = cycling['mnth'].apply(lambda x: 1 if x in [5, 6, 7, 8, 9] else 0)
cycling.peak_month.value_counts()
```

<h2>Step 4: Splitting the Data into Training and Testing Sets</h2>


```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(cycling, train_size = 0.7, test_size = 0.3, random_state = 100)
```

<h3>Rescaling the Features</h3>

```python
from sklearn.preprocessing import MinMaxScaler
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
scaler = MinMaxScaler()
num_vars =['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()

```

```python
df_train.describe()
```

```python
# Encode categorical variables with label encoding
df_train['weathersit'] = df_train['weathersit'].astype('category').cat.codes
df_train['season'] = df_train['season'].astype('category').cat.codes

plt.figure(figsize = (12,8))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()

# Calculate the correlation matrix
#corr_matrix =cycling.corr()

# Plot the heatmap
#plt.figure(figsize=(10, 8))
#sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=0.5)
#plt.title('Correlation Heatmap')
#plt.show()
```

<h2>Step 8: Making Predictions Using the Final Model</h2>

```python
num_vars =['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.describe()

```

```python
print(df_test.columns)
```

<h4>Dividing into X_test and y_test</h4>

```python
# Extract 'cnt' from df_test
y_test = df_test.pop('cnt')
X_test = df_test

# Add constant to the test data
X_test_sm = sm.add_constant(X_test)

# Drop unnecessary variables
X_test_sm = X_test_sm.drop(["atemp", "hum", "season", "workingday", 
                            "holiday", "registered", "temp", "casual"], 
                           axis=1, errors='ignore')

# Encode 'weathersit' (if categorical)
X_test_sm = pd.get_dummies(X_test_sm, columns=['weathersit'], drop_first=True)

# Align with model training features
X_test_sm = X_test_sm.reindex(columns=model.params.index, fill_value=0)

# Make predictions
y_pred_sm = model.predict(X_test_sm)

# Display predictions
print(y_pred_sm)
```

<h2>Step 9: Model Evaluation</h2>

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred_sm)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16)      
```

<h4>Residual Analysis</h4>

```python
# Calculate residuals
residuals = y_test - y_pred_sm

# Plotting residuals
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Scatter plot: Residuals vs Predicted values
ax[0].scatter(y_pred_sm, residuals, alpha=0.5)
ax[0].axhline(y=0, color='r', linestyle='--', linewidth=1)
ax[0].set_title("Residuals vs Predicted Values")
ax[0].set_xlabel("Predicted Values")
ax[0].set_ylabel("Residuals")

# Histogram: Distribution of residuals
sns.histplot(residuals, kde=True, ax=ax[1], bins=30)
ax[1].set_title("Distribution of Residuals")
ax[1].set_xlabel("Residuals")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

<h4>R squared test</h4>

```python
from sklearn.metrics import r2_score

# Assuming 'y_test' contains actual target values from the test set
# and 'y_pred_sm' contains predicted values from the test set

# Calculate R-squared score
r2 = r2_score(y_test, y_pred_sm)

# Display the result
print("R-squared Score on Test Set:", r2)
```
